A Classification Decision Tree is a machine learning algorithm used for categorical outcomes. It predicts the class of a target variable based on a series of questions about the input features. The tree is built by splitting the data into branches based on conditions that maximize the distinction between classes at each level. 

It is used for classification problems, where the goal is to assign a discrete label or category to an observation. Examples of such problems include:
-Binary Classification: Deciding between two classes (e.g., "spam" or "not spam").
-Multi-class Classification: Deciding among more than two categories (e.g., categorizing animals into "dog," "cat," "bird").

Real-World Applications:
1.Medical Diagnosis: Predicting whether a patient has a particular disease based on symptoms, test results, and demographic data.
2.Customer Segmentation: Classifying customers into segments like "loyal customers," "churn-risk customers," or "new customers."
3.Fraud Detection: Identifying transactions as "fraudulent" or "legitimate."
4.Loan Approval: Deciding whether an applicant should be granted a loan ("approve" or "reject").
5.Email Filtering: Determining whether an email is "spam" or "not spam."

b)
1.Prediction Task:
-Classification Decision Tree: Predicts categories or classes (e.g., "pass" or "fail," "red" or "blue"). It does this by recursively partitioning the data space and assigning a class label to each partition.
-Multiple Linear Regression: Predicts continuous values (e.g., house prices, stock market values) by modeling the relationship between input features and the target variable as a linear equation.

2.Method of Prediction:
-Classification Decision Tree:
  -Splits data at each node using rules based on feature thresholds (e.g., "Age > 30").
  -At each leaf node (end of a branch), it assigns the class that has the majority of observations in that node.
  -The output is categorical, based on the majority class in the terminal leaf.
  
-Multiple Linear Regression:
  -Models the relationship between the dependent variable and independent variables using a linear equation of the form:
    y = b0+b1x1+...+bnxn
     - The output is a numerical value, calculated using the weighted sum of the predictors.

3.Nature of Output:
-Classification Decision Tree: Discrete labels or probabilities for each class (e.g., "Class A with 80% confidence").
-Multiple Linear Regression: Continuous numerical values (e.g., 75.3, 42.5).

4.Model Goals:
-Classification Decision Tree: Minimize classification errors or improve the purity of class distributions at each split.
-Multiple Linear Regression: Minimize the difference between actual and predicted continuous values (typically measured by Mean Squared Error or R-squared).

```python
import pandas as pd

# Load the dataset
file_path = 'https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv'  # Update with your file path
ab = pd.read_csv(file_path, encoding="ISO-8859-1")

# Data Preprocessing
# 1) Remove Weight_oz, Width, and Height columns
ab_reduced = ab.drop(columns=["Weight_oz", "Width", "Height"], errors='ignore')

# 2) Drop all remaining rows with NaN entries
ab_reduced_noNaN = ab_reduced.dropna()

# 3) Set data types: Pub year and NumPages to int, Hard_or_Paper to category
ab_reduced_noNaN['Pub year'] = ab_reduced_noNaN['Pub year'].astype(int)
ab_reduced_noNaN['NumPages'] = ab_reduced_noNaN['NumPages'].astype(int)
ab_reduced_noNaN['Hard_or_Paper'] = ab_reduced_noNaN['Hard_or_Paper'].astype('category')

# Save or display the processed dataset
print(ab_reduced_noNaN)
```

```python
from sklearn.model_selection import train_test_split

# Splitting the data
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(ab_reduced_noNaN, test_size=0.2, random_state=42)

# Reporting the number of observations
print(f"Training dataset observations: {len(ab_reduced_noNaN_train)}")
print(f"Test dataset observations: {len(ab_reduced_noNaN_test)}")
```

```python
# Splitting the data using df.sample
train_size = int(len(ab_reduced_noNaN) * 0.8)
ab_reduced_noNaN_train = ab_reduced_noNaN.sample(train_size, random_state=42)
ab_reduced_noNaN_test = ab_reduced_noNaN.drop(ab_reduced_noNaN_train.index)

# Reporting the number of observations
print(f"Training dataset observations: {len(ab_reduced_noNaN_train)}")
print(f"Test dataset observations: {len(ab_reduced_noNaN_test)}")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

# Step 1: Preparing the data
y = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']  # Hard cover as 1, Paper back as 0
X = ab_reduced_noNaN_train[['List Price']]

# Step 2: Initializing the classifier
clf = DecisionTreeClassifier(max_depth=2, random_state=42)

# Step 3: Training the classifier
clf.fit(X, y)

# Step 4: Plotting the tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
tree.plot_tree(clf, feature_names=['List Price'], class_names=['Paper Back', 'Hard Cover'], filled=True)
plt.show()
```

The tree.plot_tree(clf) visualization will show splits based on the List Price variable:

Each node represents a decision split on List Price.
Leaf nodes represent the predicted class (hard cover or paper back).
The depth of the tree stops at 2 levels due to max_depth=2.
Predictions are made by traversing from the root to a leaf based on whether a given List Price value satisfies the splitting conditions at each node.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Step 1: Preparing the data
y = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']  # Hard cover as 1, Paper back as 0
X = ab_reduced_noNaN_train[['NumPages', 'Thick', 'List Price']]

# Step 2: Initializing the classifier
clf2 = DecisionTreeClassifier(max_depth=4, random_state=42)

# Step 3: Training the classifier
clf2.fit(X, y)

# Step 4: Plotting the decision tree
plt.figure(figsize=(16, 10))
tree.plot_tree(clf2, feature_names=['NumPages', 'Thick', 'List Price'], class_names=['Paper Back', 'Hard Cover'], filled=True)
plt.show()
```

1)Tree Structure:
-Each internal node represents a decision rule based on one of the features (NumPages, Thick, List Price).
-At each node, the data is split into two groups based on the condition, minimizing the classification error.

2)Leaf Nodes:
-Each leaf node corresponds to a final decision, predicting either "Paper Back" (class 0) or "Hard Cover" (class 1).
-Leaf nodes also display the proportion of samples belonging to each class and the total number of samples reaching that node.

3)Prediction Process:
-Starting from the root node, a book's feature values are compared to the decision thresholds at each split.
-The traversal continues down the tree until a leaf node is reached, where the prediction is made based on the majority class.

4)Depth Limitation:
-With max_depth=4, the tree can make up to 4 splits from the root node to a leaf.
-This limits the complexity of the model and reduces the risk of overfitting.

```python
from sklearn.metrics import confusion_matrix, accuracy_score

# Prepare the test dataset
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']  # Target variable
X_test_clf = ab_reduced_noNaN_test[['List Price']]                    # Features for clf
X_test_clf2 = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]  # Features for clf2

# Generate predictions
y_pred_clf = clf.predict(X_test_clf)
y_pred_clf2 = clf2.predict(X_test_clf2)

# Confusion matrices
conf_matrix_clf = confusion_matrix(y_test, y_pred_clf)
conf_matrix_clf2 = confusion_matrix(y_test, y_pred_clf2)

# Function to calculate sensitivity, specificity, and accuracy
def calculate_metrics(conf_matrix):
    tn, fp, fn, tp = conf_matrix.ravel()
    sensitivity = tp / (tp + fn)  # Recall
    specificity = tn / (tn + fp)
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    return sensitivity, specificity, accuracy

# Calculate metrics for clf
sensitivity_clf, specificity_clf, accuracy_clf = calculate_metrics(conf_matrix_clf)

# Calculate metrics for clf2
sensitivity_clf2, specificity_clf2, accuracy_clf2 = calculate_metrics(conf_matrix_clf2)

# Print results
print("Confusion Matrix for clf (List Price):\n", conf_matrix_clf)
print(f"Sensitivity: {sensitivity_clf:.2f}, Specificity: {specificity_clf:.2f}, Accuracy: {accuracy_clf:.2f}\n")

print("Confusion Matrix for clf2 (NumPages, Thick, List Price):\n", conf_matrix_clf2)
print(f"Sensitivity: {sensitivity_clf2:.2f}, Specificity: {specificity_clf2:.2f}, Accuracy: {accuracy_clf2:.2f}")
```

2)Proportion of true negatives correctly identified.

Specificity=(TN)/(TN+FP)
